```python
# Dependencies
import pandas as pd
```

```python
# Store filepath in a variable
csv_file = "Resources/cdc-aids-deaths-by-state.csv"
```

```python
# Read our Data file with the pandas library
# Not every CSV requires an encoding, but be aware this can come up
file_df = pd.read_csv(csv_file, encoding="ISO-8859-1")
```

```python
file_df
```

```python
grouped_state = file_df.groupby("geography")

#grouped_ind_column = grouped_state["indicator"]

grouped_geo_column = grouped_state["geography"]

state_avg_fips = round(grouped_state["fips"].mean(),2)

#grouped_race_column = grouped_state["race"]

#grouped_sex_column = grouped_state["sex"]

#grouped_age_column = grouped_state["age_group"]

#grouped_misc_column = grouped_state["misc"]

state_avg_rate = round(grouped_state["rate"].mean(),2)

state_avg_cases = round(grouped_state["cases"].mean(),2)

state_avg_population = round(grouped_state["population"].mean(),2)

#Create dataframe and rename column names with dictionary
clean_file_df = pd.DataFrame({#"indicator":grouped_ind_column, 
                                  "fips":state_avg_fips, #"race":grouped_race_column,
                                  #"sex":grouped_sex_column, "age_group":grouped_age_column,
                                  #"misc":grouped_misc_column,
                                  "avg_rate": state_avg_rate,
                                  "avg_cases": state_avg_cases, "avg_population": state_avg_population
                                  })
clean_file_df = clean_file_df.reset_index()
clean_file_df
```

```python
# Export file as a CSV, without the Pandas index, but with the header
clean_file_df.to_csv("Output/2000-2014_cdc-aids-deaths-by-state_averages.csv", index=False, header=True)
```
