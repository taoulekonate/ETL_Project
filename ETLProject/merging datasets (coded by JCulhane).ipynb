```python
# Dependencies
import pandas as pd
```

```python
# Store filepath in a variable
csv_file1 = "Resources/new_providers.csv"
csv_file2 = "Output/2000-2014_cdc-aids-deaths-by-state_averages.csv"
```

```python
# Read our Data file with the pandas library
# Not every CSV requires an encoding, but be aware this can come up
file1_df = pd.read_csv(csv_file1, encoding="ISO-8859-1")
file2_df = pd.read_csv(csv_file2, encoding="ISO-8859-1")
```

```python
# Modules
import os
import csv

# Set path for file
csvpath1 = os.path.join("Resources", "hiv_providers.csv")
csvpath2 = os.path.join("Output", "2000-2014_cdc-aids-deaths-by-state_averages.csv")

statelist = []

with open(csvpath2, newline="") as csvfile:
    csvreader = csv.reader(csvfile, delimiter=",")
    for row in csvreader:
        statelist.append(row[0])

stateset = set(statelist)


# Open the CSV
with open(csvpath1, newline="") as csvfile:
    csvreader = csv.reader(csvfile, delimiter=",")

    # Loop through looking for the video
    for row in csvreader:
        splitrow = row[2].split(" ")
        splitrow.pop(0)
        splitset = set(splitrow)
        inter = stateset.intersection(splitset)
        if inter != "":
            print(inter)
```

```python
merged1_df = pd.merge(file1_df, file2_df, how="left", on=['geography', 'geography'])

file1_grp = file1_df.groupby("geography")
total_clinics = file1_grp["id"].count()
counted = pd.DataFrame({"Total Clinics":total_clinics})
counted = counted.reset_index()
counted
merged2_df = pd.merge(counted, file2_df, how="left", on=['geography', 'geography'])
```

```python
# Export file as a CSV, without the Pandas index, but with the header
merged1_df.to_csv("Output/2000-2014_with_clinic_names.csv", index=False, header=True)
merged2_df.to_csv("Output/2000-2014_states_by_number_of_clinics_and_HIV_deaths.csv", index=False, header=True)
```
